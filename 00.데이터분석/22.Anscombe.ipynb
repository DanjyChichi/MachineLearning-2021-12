Anscombe's Quartet

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ans.shape
```

- dataset I 에 대해서 그래프를 그려봄

```python
ds = ans[ans.dataset == 'I']
ds
```

- 회귀식 y = 0.5 * x + 3

```python
plt.scatter(ds.x, ds.y)
plt.plot([4,14], [5,10], 'r-', lw=2)
plt.title('Dataset I')
plt.show()
```

- 최종 그림

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8), sharex=True, sharey=True)
for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ax = axs[i//2, i%2]
    ds = ans[ans.dataset == roman]
    ax.scatter(ds.x, ds.y, s=50)
    ax.plot([3,20], [4.5,13], 'r-', lw=2)
    ax.set_ylim(2,13)
    ax.set_title(f'Dataset {roman}')
plt.suptitle("Anscombe's Quartet", fontsize=20)
plt.tight_layout()
plt.show()
```
